```python
import numpy as np
import matplotlib.pyplot as plt

# set x 
x = range(400) 

# set y as exponential of x
y =[val**2 for val in x] 

# Plot x vs. y
plt.plot(x,y)
```

```python
# Plot with two series
x = np.linspace(0, 3*np.pi, 100)
y =np.sin(x)
z = np.cos(x)

plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
new_list = [[5., 25., 50., 20.], 
            [9., 31., 43., 28.], 
            [4., 23., 51., 17.], 
            [6., 22., 52., 19.]]

x = np.arange(4) 

plt.bar(x + 0.0, new_list[0], color ='b', width =0.2)
plt.bar(x + 0.2, new_list[1], color ='r', width =0.2)
plt.bar(x + 0.4, new_list[2], color ='g', width =0.2)
plt.bar(x + 0.6, new_list[3], color ='orange', width =0.2)

plt.show()
```

```python
data_a = np.array([5., 30., 45., 22.]) 
data_b = np.array([5., 25., 50., 20.])
z = np.arange(4)
# - notation is needed for generating back to back charts
plt.barh(z, data_a, color ='orange')
plt.barh(z, -data_b, color ='blue')
```
